```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://www.flipkart.com/womens-dresses/tokyo-talkies~brand/pr?sid=clo%2Codx%2Cmaj%2Cjhy" 
```

```python
allow = requests.get(url)
```

```python
print (allow)
```

```python
soup=BeautifulSoup(allow.text)
```

```python
dresses_name=soup.select(".IRpwTa")
```

```python
dresses_name
```

```python
dresses=[]
dresses_links=[]
```

```python
for item in dresses_name:
    dresses.append(item.text)
    link = "http://www.flipkart.com"
    dresses_links.append(link)
```

```python
dresses
```

```python
dresses_rates=soup.select("._3bPFwb ._30jeq3")
```

```python
dresses_rates
```

```python
rates=[]
```

```python
for item in dresses_rates:
    rates.append(item.text)
```

```python
rates
```

```python
import pandas as pd
```

```python
dresses = pd.DataFrame (dresses_name,columns=['dresses'])
```

```python
dresses ['rates'] = rates
```

```python
dresses
```

```python
dresses.to_csv('dresses.csv')
```
